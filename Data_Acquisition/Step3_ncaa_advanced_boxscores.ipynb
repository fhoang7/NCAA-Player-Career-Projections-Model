# Goal:
Take the simple NCAA Boxscore and add the advanced statistics into it.

### Libraries
We are bringing the necessary libraries in order to access and process the data.

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sportsreference.ncaab.teams import Teams
from sportsreference.ncaab.roster import Roster, Player
from sportsreference.ncaab.boxscore import Boxscore, Boxscores
pd.set_option('display.max_columns', 500)
```

Read in the simple boxscores from the source folder. Examine the data to ensure it is correct.

```python
ncaa_boxscores = pd.read_csv('../Data/ncaa_simple_boxscores.csv')
```

```python
ncaa_boxscores.shape
```

```python
ncaa_boxscores.tail()
```

Create a dataframe without null boxscores.

```python
nb = ncaa_boxscores.iloc[53189:,:]
```

```python
nb = nb.loc[nb.boxscore.notnull()]
```

```python
nb.shape
```

With the IDs from boxscores with no null data, run a for loop to join the advanced data to these boxscores.

```python
for n in nb.boxscore:
    game = Boxscore(n)
    box = game.dataframe.reset_index().rename(columns={'index':'boxscore'})
    ncaa_advanced_boxscores = pd.concat([ncaa_advanced_boxscores,box])
```

Examine the shape and the structure of the dataframes created.

```python
ncaa_advanced_boxscores.shape
```

```python
ncaa_advanced_boxscores.head()
```

```python
ncaa_advanced_boxscores.tail()
```

We notice that the precentages are not how we want them displayed. They are inbetween 0 - 100, while we are looking for them to be between 0 - 1. To address this we simply divide the precentage columns by 100 to get them into the structure we desire. 

```python
ncaa_advanced_boxscores.away_assist_percentage = ncaa_advanced_boxscores.away_assist_percentage / 100
ncaa_advanced_boxscores.away_block_percentage = ncaa_advanced_boxscores.away_block_percentage / 100
ncaa_advanced_boxscores.away_defensive_rebound_percentage = ncaa_advanced_boxscores.away_defensive_rebound_percentage / 100
ncaa_advanced_boxscores.away_offensive_rebound_percentage = ncaa_advanced_boxscores.away_offensive_rebound_percentage / 100
ncaa_advanced_boxscores.away_steal_percentage = ncaa_advanced_boxscores.away_steal_percentage / 100
ncaa_advanced_boxscores.away_total_rebound_percentage = ncaa_advanced_boxscores.away_total_rebound_percentage / 100
ncaa_advanced_boxscores.away_turnover_percentage = ncaa_advanced_boxscores.away_turnover_percentage / 100
```

```python
ncaa_advanced_boxscores.home_assist_percentage = ncaa_advanced_boxscores.home_assist_percentage / 100
ncaa_advanced_boxscores.home_block_percentage = ncaa_advanced_boxscores.home_block_percentage / 100
ncaa_advanced_boxscores.home_defensive_rebound_percentage = ncaa_advanced_boxscores.home_defensive_rebound_percentage / 100
ncaa_advanced_boxscores.home_offensive_rebound_percentage = ncaa_advanced_boxscores.home_offensive_rebound_percentage / 100
ncaa_advanced_boxscores.home_steal_percentage = ncaa_advanced_boxscores.home_steal_percentage / 100
ncaa_advanced_boxscores.home_total_rebound_percentage = ncaa_advanced_boxscores.home_total_rebound_percentage / 100
ncaa_advanced_boxscores.home_turnover_percentage = ncaa_advanced_boxscores.home_turnover_percentage / 100
```

Re-examine the dataframe to make sure it is the way we want it and then save to CSV.

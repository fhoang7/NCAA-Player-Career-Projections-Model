When errors occure use the last print out as the new starting length in the loop.

```python
for n in range(242893 + 1):
    print(242893 - n)
    game_data = Boxscore(ncaa_boxscores.boxscore[n])
    home_players = game_data.home_players
    away_players = game_data.away_players
    for home_player in home_players:
        player_data = home_player.dataframe
        player_data = player_data.reset_index().rename(columns={'index':'player_id'})
        player_data['boxscore'] = ncaa_boxscores.boxscore[n]
        player_data['team_abbr'] = ncaa_boxscores.home_abbr[n]
        player_data['venue'] = 'Home'
        ncaa_player_boxscores = pd.concat([ncaa_player_boxscores, player_data])
    for away_player in away_players:
        player_data = away_player.dataframe
        player_data = player_data.reset_index().rename(columns={'index':'player_id'})
        player_data['boxscore'] = ncaa_boxscores.boxscore[n]
        player_data['team_abbr'] = ncaa_boxscores.away_abbr[n]
        player_data['venue'] = 'Away'
        ncaa_player_boxscores = pd.concat([ncaa_player_boxscores, player_data])
    if n in range(100, 242893, 100):
        ncaa_player_boxscores.to_csv('../Data/ncaa_player_boxscores.csv', index = False)
```

Write the new dataframe to a csv.

```python
ncaa_player_boxscores.to_csv('../Data/ncaa_player_boxscores.csv',index=False)
```

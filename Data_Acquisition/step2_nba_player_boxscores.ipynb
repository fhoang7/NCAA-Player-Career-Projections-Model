```python
for n in range(61241 + 1):
    print(61241 - n)
    game_data = Boxscore(boxscores.boxscore[n])
    home_players = game_data.home_players
    away_players = game_data.away_players
    for home_player in home_players:
        player_data = home_player.dataframe
        player_data = player_data.reset_index().rename(columns={'index':'player_id'})
        player_data['boxscore'] = boxscores.boxscore[n]
        player_data['team_abbr'] = boxscores.home_abbr[n]
        player_data['venue'] = 'Home'
        player_boxscores = pd.concat([player_boxscores,player_data])
    for away_player in away_players:
        player_data = away_player.dataframe
        player_data = player_data.reset_index().rename(columns={'index':'player_id'})
        player_data['boxscore'] = boxscores.boxscore[n]
        player_data['team_abbr'] = boxscores.away_abbr[n]
        player_data['venue'] = 'Away'
        player_boxscores = pd.concat([player_boxscores,player_data])
    if n % 500 == 0:
        player_boxscores.to_csv('../Data/nba_player_boxscores.csv',index=False)
```

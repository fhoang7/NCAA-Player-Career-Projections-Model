# Goal: 
Scrape the entirety of basic box score data exposed by the SportsReference API. 


Here, we import the libraries we need in order to perform data exploration and carpentry on our nba box score data. The Sports Reference package allows us to interact with the data. 

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sportsreference.nba.teams import Teams
from sportsreference.nba.roster import Roster, Player
from sportsreference.nba.boxscore import Boxscore, Boxscores
pd.set_option('display.max_columns', 500)
```

```python
basic_boxscores = pd.DataFrame()
```

Here, we are loading boxscores from 1940 to 2019. We are extracting box score data by season (defined as starting in August of a particular year and concludes in July of the next year). Games for a particular day are packed inside the games suffix so we must call games.games in order to unpack the data. We then iterate through each game for each day in the NBA seasons and load the boxscores into the `basic_boxscores` dataframe. 

After populating the dataframe, we then write it to CSV for future usage.

```python
for year in range(1982,1945,-1):
    print(year)
    games = Boxscores(datetime(year-1,8,1),datetime(year,7,30))
    games_dict = games.games
    for date, games in games_dict.items():
        for game in games:
            cols = game.keys()
            cols = list(cols)
            cols.append('season')
            data = pd.DataFrame(game.values()).transpose()
            data['season'] = year
            basic_boxscores = pd.concat([basic_boxscores,data])
    print(year,' Loaded\n=================')
    basic_boxscores.to_csv('../Data/nba_boxscores.csv', index = False)
```

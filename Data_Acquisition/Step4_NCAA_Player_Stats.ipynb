```python
for year in range(2020,1900,-1):
    print(year)
    if year < 2011:
        for team in team_names:
            print(team)
            try:
                roster = Roster(team,year=year)
                for player in roster.players:
                    print(player.name)
                    players.append(player.player_id)
            except:
                print('Missing data for: {} {}'.format(year, team))
                missing_years.append(year)
                missing_teams.append(team)
    else:
        teams = Teams(year=year)
        for team in teams:
            print(team.name)
            try:
                roster = team.roster
                for player in roster.players:
                    print(player.name)
                    players.append(player.player_id)
            except:
                print('Missing data for: {} {}'.format(year, team))
                missing_years.append(year)
                missing_teams.append(team)
```

Use the `set()` funtion to whittle the players list down to the set of players, so that each value only occurs once, and then convert back into a list.

Create an empty dataframe called `player_career_stats` that will be used to store the results of the for loop in the next cell of code.

```python
player_career_stats = pd.DataFrame()
```

These for loops use the sports reference library `Player()` funtion to generate player statistics data and also accomplish a few data carpentry tasks, such as renaming columns and dropping certain rows.  All of this data is concatenated into the `player_career_stats` dataframe.

```python
for n in range(0,len(player_ids)):
    try:
        player_info = Player(player_ids[n])
        df = player_info.dataframe
        df = df.reset_index().rename(columns={'level_0':'season'})
        df = df.loc[df.season!='Career']
        years = []
        for season in df['season']:
            years.append(int(season.split('-')[0])+1)
        df.season = years
        df['player_name'] = player_info.name
        print(player_info.name)
        player_career_stats = pd.concat([player_career_stats,df])
    except:
        pass
```

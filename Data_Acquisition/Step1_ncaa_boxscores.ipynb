Import necesary libraries

```python
! pip install sportsreference
import pandas as pd
import numpy as np
from datetime import datetime
from sportsreference.ncaab.teams import Teams
from sportsreference.ncaab.roster import Roster, Player
from sportsreference.ncaab.boxscore import Boxscore, Boxscores
pd.set_option('display.max_columns', 500)
```

Create an empty dataframe that will hold NCAA game by game boxscore data

```python
ncaa_boxes = pd.DataFrame()
```

This code makes use of the `Boxcores()` function from the sport reference package to pull boxscore data using the sport reference api.  This data gets stored as dictionary.  The for loops decompose this dictionary and transform it into a dataframe, storing it the `nccaa_boxes dataframe` created in the previous step.  A list of column names is also created in this loop. 

```python
for year in range(2019,1889,-1):
    print(year)
    games = Boxscores(datetime(year-1,7,1),datetime(year,6,30))
    games_dict = games.games
    for date, games in games_dict.items():
        for game in games:
            cols = game.keys()
            cols = list(cols)
            cols.append('season')
            data = pd.DataFrame(game.values()).transpose()
            data['season'] = year
            ncaa_boxes = pd.concat([ncaa_boxes,data])
    ncaa_boxes.to_csv('../Data/ncaa_simple_boxscores.csv',index=False)
    print(year,' Loaded\n====================')
```

Rename the columns in `ncaa_boxes` using the list of coulmn names created in the previous step.  

```python
ncaa_boxes.columns = cols
```

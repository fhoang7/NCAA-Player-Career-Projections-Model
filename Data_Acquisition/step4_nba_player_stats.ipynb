### Libraries

```python
! pip install sportsreference
import pandas as pd
import numpy as np
from datetime import datetime
from sportsreference.nba.teams import Teams
from sportsreference.nba.roster import Roster, Player
from sportsreference.nba.boxscore import Boxscore, Boxscores, BoxscorePlayer, AbstractPlayer
from sportsreference.nba.schedule import Schedule, Game
pd.set_option('display.max_columns', 500)
```

Read in players box scores from source folder.

```python
player_boxes = pd.read_csv('../Data/nba_player_boxscores.csv')
```

Sort out the unique player IDs to not have duplicated data. 

```python
player_ids = player_boxes.player_id.unique()
```

```python
len(player_ids)
```

```python
players = []
missing_years = []
missing_teams = []
```

Loop through every NBA team from 1940-2019 and assemble a dataframe of player names. If player names are missing, add the missing player's team and year to the `missing_teams` and `missing_years` lists for accountability. 

```python
for year in range(2019,1940,-1):
    print(year)
    teams = Teams(year=year)
    for team in teams:
        print(team.name)
        try:
            roster = team.roster
            for player in roster.players:
                print(player.name)
                players.append(player.player_id)
        except:
            print('Missing data for: {} {}'.format(year, team.name))
            missing_years.append(year)
            missing_teams.append(team.name)
```

```python
player_ids = set(players)
player_ids = list(player_ids)
```

```python
len(player_ids)
```

```python
player_career_stats = pd.DataFrame()
```

Use the unique player IDs to use in the sportsreference library. The library will give the data for each player, and that is appended in this for loop. 

```python
for n in range(len(player_ids)):
    player_info = Player(player_ids[n])
    df = player_info.dataframe
    df = df.reset_index().rename(columns={'level_0':'season'})
    df = df.loc[df.season!='Career']
    years = []
    for season in df['season']:
        years.append(int(season.split('-')[0])+1)
    df.season = years
    df['player_name'] = player_info.name
    print(player_info.name)
    player_career_stats = pd.concat([player_career_stats,df])
```
